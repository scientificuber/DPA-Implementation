```python
guessed_byte = 0

max_diffs = [0]*256
full_diffs = [0]*256


for guess in tnrange(0, 256):
    
    one_list = []
    zero_list = []
    
    for trace_index in range(numtraces):
        
        state = np.bitwise_xor(guess, ciphers[trace_index][guessed_byte])
        s = INV_SBOX[state]
        val = np.bitwise_xor(ciphers[trace_index][guessed_byte],s)
        if val & (2**7):        
            one_list.append(trace_array[trace_index][2340:2395])
        else:
            zero_list.append(trace_array[trace_index][2340:2395])
            
    one_avg = np.asarray(one_list).mean(axis=0)
    zero_avg = np.asarray(zero_list).mean(axis=0)
    full_diffs[guess] = abs(one_avg - zero_avg)
    max_diffs[guess] = np.max(full_diffs[guess])
    
#     print("Guessing %02x: %f"%(guess, mean_diffs[guess]))

sorted_args = np.argsort(max_diffs)[::-1]
```

```python
def mean(X):
    return np.sum(X, axis=0)/len(X)

def std_dev(X, X_bar):
    return np.sqrt(np.sum((X-X_bar)**2, axis=0))

def cov(X, X_bar, Y, Y_bar):
    return np.sum((X-X_bar)*(Y-Y_bar), axis=0)
```

```python
HW = [bin(n).count("1") for n in range(0, 256)]
```

```python
guessed_byte = 0

max_cpa = [0]*256

t_bar = mean(trace_array[:,2340:2395]) 
o_t = std_dev(trace_array[:,2340:2395], t_bar)

for guess in tnrange(0, 256):
    
    hw = []
    
    for trace_index in range(numtraces):
        
        state = np.bitwise_xor(guess, ciphers[trace_index][guessed_byte])
        s = INV_SBOX[state]
        val = np.bitwise_xor(ciphers[trace_index][guessed_byte],s)
        hw.append(HW[val])
        
    hws = np.array(hw).reshape((numtraces,1))
#     print(type(hws))
    
    hws_bar = mean(hws)
#     print(hws.shape,hws_bar.shape,t_bar.shape,trace_array[:,2340:2395].shape)
#     break
    o_hws = std_dev(hws, hws_bar)
    correlation = cov(trace_array[:,2340:2395], t_bar, hws, hws_bar)
    cpaoutput = correlation/(o_t*o_hws)
    max_cpa[guess] = max(abs(cpaoutput))
    
#     print("Guessing %02x: %f"%(guess, mean_diffs[guess]))

sorted_args = np.argsort(max_cpa)[::-1]
```
