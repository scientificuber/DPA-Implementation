```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"]=20,5
```

```python
dataset = pd.read_csv("./dpav2_trace_files1/wave_aist-aes-agilent_2009-12-30_22-53-19_n=0_k=0000000000000003243f6a8885a308d3_m=0000000000000002b7e151628aed2a6a_c=e6a636e30c85f35e980f3546a04daff7.csv")
dataset.head()
```

```python
headers = dataset.head(23)
headers
```

```python
data = dataset.tail(3253)
data.head(5)
```

```python
# convert the input object to int format to plot
data.dtypes
data = data.astype(str).astype(int)
data.dtypes
```

```python
# Plotting the trace in its entirety.
# A trace lookup could give important information about the structure of the underlying implementation

plt.plot(data)
plt.xlabel('time', fontsize=12)
plt.ylabel('current', fontsize=12)
plt.title('Trace Data 1', fontsize=13)
plt.ylim(-20000, 20000)
plt.show() 


```

```python
dataset2 = pd.read_csv("./dpav2_trace_files1/wave_aist-aes-agilent_2009-12-30_22-53-19_n=1_k=0000000000000003243f6a8885a308d3_m=bf7158809cf4f3c762e7160f38b4da56_c=ff869d789be05b4b673662de27dd1bcd.csv")
data2 = dataset2.tail(3253)
data2.dtypes
data2 = data2.astype(str).astype(int)
data2.dtypes

plt.plot(data2)

plt.xlabel('time', fontsize=12)
plt.ylabel('current', fontsize=12)
plt.title('Trace Data 2', fontsize=13)
plt.ylim(-20000, 20000)

plt.show()
```

This is AES-128 implementation(unmasked)

Note that from 450 we see a repeating structure that continues till 2400

There are 9 similar structures within 450-2400.They correspond to the first 9 round(Key XOR, SBox lookup). The last round is expected to be dissimilar.(Key XOR, SBox lookup, Last Key XOR)
